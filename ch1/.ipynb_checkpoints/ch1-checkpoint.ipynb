```python
from src.Mean_EBinning.Mean_EBinning import Mean_EBinning
import json
from scipy.stats import zscore
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from src.utility.utility_kepler_flare import default_kepler_flare
from statsmodels.tsa.seasonal import STL
import os
```

## Figure 1.1

```python
TEMPLATE_FILE_NAME = os.path.expanduser("~/source_code/thesis_sourcecode/datasets/clean_lc")
FLARE_HEIGHT = 1.5
# Set starting and ending point of flare
STARTING_POINT = 200

kepler_flare = default_kepler_flare(FLARE_HEIGHT)

sample_data = []
raw_data = []
with open(TEMPLATE_FILE_NAME, "r") as sample_file:
    for line in sample_file:
        sample_data.append(float(line))
        raw_data.append(float(line))

sample_data = zscore(np.array(sample_data))
raw_data = zscore(np.array(raw_data))


sample_data[STARTING_POINT: STARTING_POINT+len(kepler_flare)] = sample_data[STARTING_POINT: STARTING_POINT+len(kepler_flare)] + kepler_flare


# plt.plot(sample_data[:1000])
# plt.plot(sketching_result[:1000], color='r')
# plt.axvline(STARTING_POINT, color='green', alpha=0.5)
# plt.axvline(STARTING_POINT+len(kepler_flare), color='green', alpha=0.5)
# plt.show()

file_name = os.path.expanduser("~/source_code/thesis_sourcecode/datasets/oisst2.1_world2_sst_day.json")

# Load JSON data
with open(file_name, 'r') as file:
    json_data = json.load(file)

ts = []
for sublist in json_data[1:42]:
    ts += [element for element in sublist['data'] if element != None]



ts = ts[-8000:]

start_date = '1982-01-01'
end_date = '2022-12-31'
date_sequence = pd.date_range(start=start_date, end=end_date)
date_sequence = date_sequence[-8000:]
df = pd.DataFrame({'Date': date_sequence, 'Data': ts})

fig, axs = plt.subplots(2, 1, figsize=(12, 7))
axs[0].plot(sample_data[:1000], label ='Raw data')
axs[0].set_title('Light curve data', fontsize=15)
axs[0].set_xlabel('Timestamps', fontsize=15)
axs[0].set_ylabel('Normalized flux', fontsize=15)


axs[1].plot(df['Date'], df['Data'])
axs[1].set_title('Daily sea surface temperature', fontsize=15)
axs[1].set_xlabel('Date', fontsize=15)
axs[1].set_ylabel('Temperature (°C)', fontsize=15)

# Adjust the layout to prevent overlap
plt.tight_layout()
```

## Figure 1.2

```python
TEMPLATE_FILE_NAME = os.path.expanduser("~/source_code/data_stream/thesis/dataset/clean_lc")
FLARE_HEIGHT = 1.5
# Set starting and ending point of flare
STARTING_POINT = 200

kepler_flare = default_kepler_flare(FLARE_HEIGHT)

sample_data = []
raw_data = []
with open(TEMPLATE_FILE_NAME, "r") as sample_file:
    for line in sample_file:
        sample_data.append(float(line))
        raw_data.append(float(line))

sample_data = zscore(np.array(sample_data))
raw_data = zscore(np.array(raw_data))


sample_data[STARTING_POINT: STARTING_POINT+len(kepler_flare)] = sample_data[STARTING_POINT: STARTING_POINT+len(kepler_flare)] + kepler_flare

EBinning = Mean_EBinning(max_window_size=10, ini_binsize=12, alpha=0.95)

for instance in sample_data:
    EBinning.add_element(instance)

EBinning.insert_latest_buffer()

sketching_result = []
for bin_list in EBinning.get_window:
    temp = [bin_list.get_mu] * bin_list.get_n
    sketching_result = sketching_result + temp

file_name = os.path.expanduser("~/source_code/data_stream/thesis/dataset/oisst2.1_world2_sst_day.json")
with open(file_name, 'r') as file:
    json_data = json.load(file)

ts = []
for sublist in json_data[1:42]:
    ts += [element for element in sublist['data'] if element != None]

stl = STL(ts, period=365, robust=True).fit()

seasonal = pd.Series(stl.seasonal[-8000:])

start_date = '1982-01-01'
end_date = '2022-12-31'
date_sequence = pd.date_range(start=start_date, end=end_date)
date_sequence = date_sequence[-8000:]
df = pd.DataFrame({'Date': date_sequence, 'Data': seasonal})

fig, axs = plt.subplots(2, 1, figsize=(12, 7))
axs[0].plot(sample_data[:1000], label ='Raw data', alpha = 0.5)
axs[0].plot(sketching_result[:1000], label ='sketching result', color='red')
axs[0].axvline(STARTING_POINT, color='g', alpha=0.7)
axs[0].axvline(STARTING_POINT + len(kepler_flare), color='g', alpha=0.7)
axs[0].set_title('Light curve data', fontsize=15)
axs[0].set_xlabel('Timestamps', fontsize=15)
axs[0].set_ylabel('Normalized flux', fontsize=15)
axs[0].legend(fontsize="13")

axs[1].plot(df['Date'], df['Data'])
axs[1].set_title('Daily sea surface temperature', fontsize=15)
axs[1].set_xlabel('Date', fontsize=15)
axs[1].set_ylabel('Temperature (°C)', fontsize=15)


plt.tight_layout()
```
