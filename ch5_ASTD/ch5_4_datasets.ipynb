```python
import matplotlib.pyplot as plt
import os
import json
import pandas as pd
import src.ASTD.ECML_synthetic as gen_syn
```

```python
data1 = gen_syn.generate_syn1(is_export = False)
fig, axes = plt.subplots(2, 2, figsize=(14, 6), sharex=True)
axes = axes.flatten()
title = ['Time series data (Y)', 'Trend component (T)', 'Seasonal component (S)', 'Residual component (R)']
axes[0].plot(data1['ts'], label='ts')
axes[1].plot(data1['trend'], label='trend')
axes[2].plot(data1['seasonal'], label='seasonal')
axes[3].plot(data1['residual'], label='residual')
min_y = min(data1['ts'])
max_y = max(data1['ts'])
# Set common x and y ticks
for i, ax in enumerate(axes):
    ax.tick_params(axis='both', labelsize=18)
    ax.set_xlabel('Timestamps', fontsize=18)
    ax.set_ylabel('Value', fontsize=18)
    ax.set_title(title[i], size=18)
    if i in [0,1]:
        ax.set_ylim(min_y, max_y)
        
# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
